```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("Netflix.csv")
```

```python
df.head(5)
```

```python
   # obj 1 -> Volume of stock traded
```

```python
df['Date'] = pd.to_datetime(df['Date'])            # setting date as the index and import real date and time 
df = df.set_index('Date')
df.head()
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
```

```python
sns.lineplot(x = df.index , y = df['Volume'], label = 'Volume')       # Visualization of stock w.r.to time 
plt.title('Volume & Stock    Vs  Time' )
```

```python
  # obj2 = Netflix stock price - High , Open , CLose -
```

```python
df.plot(y = ['Open','High','Close' ], ylabel = 'price'  , title = "Netflix Stock at opening,high,closing in (USD) " )
```

```python
# obj3 -  Netflix stock price - Day,Monthly,Year wise
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize = (15,10))
df.groupby(df.index.day).mean().plot(y = "Volume", ax = ax1, ylabel= 'Price' , xlabel = 'DAY')
df.groupby(df.index.month).mean().plot(y = "Volume", ax = ax2, ylabel= 'Price' , xlabel = 'Month')
df.groupby(df.index.year).mean().plot(y = "Volume", ax = ax3, ylabel= 'Price' , xlabel = 'Year')
```
